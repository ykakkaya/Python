```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}        
)

df
```

# sayısal seğişkenlerde atama
* ilk önce 0 atayalım
* ikinci adımda ortalamayı atayalım

```python
df["V1"].fillna(0)
```

```python
df["V1"].fillna(df["V1"].mean())
```

* tüm değişkenleri kendi ort. ile dolduralım

```python
df.apply(lambda x: x.fillna(x.mean(),axis=0))
```

* ikinci yol

```python
df.fillna(df.mean()[:])
```

```python
df.fillna(df.mean()["V1":"V2"])
```

```python
df["V3"].fillna(df["V3"].median())
```

* üçüncü yol

```python
df.where(pd.notna(df), df.mean(), axis = "columns")
```
