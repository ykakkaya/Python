```python
import seaborn as sns
flights=sns.load_dataset("flights")
df=flights.copy()
```

```python
df.head(3)
```

```python
df.info()
```

```python
sns.barplot(x="year",y="passengers",data=df)
```

```python
sns.boxplot(x="year",y="passengers",hue="month",data=df)
```

```python
sns.catplot(x="year",y="passengers",hue="month",data=df)
```

```python
sns.lmplot(x="year",y="passengers",hue="month",data=df)
```

```python
af=df.pivot(index="year",columns="month",values="passengers")
```

```python
af
```

```python
sns.heatmap(af)
```

```python
sns.heatmap(af,annot=True,fmt="d",linewidths=1)
```
