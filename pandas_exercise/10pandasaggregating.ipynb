```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"gruplar":["a","b","c","a","b","c"],
                "degiskenler1":[10,23,33,22,11,99],
                "degiskenler2":[100,253,333,262,111,969]},columns=["gruplar","degiskenler1","degiskenler2"])
df
```

# aggregate, filter, transform, apply 

```python
df.groupby("gruplar").mean()
```

## _aggregate fonk. kullanılırken listeye kullanmak istediğimiz fonk. yazılır._

```python
df.groupby("gruplar").aggregate(["count","min","max","mean","std","var",np.median])
```

### her değişken için farklı işlemler yaptırabiliriz.

```python
df.groupby("gruplar").aggregate({"degiskenler1":["count","min","max"]
                                ,"degiskenler2":["std","var","mean",np.median]})
```
