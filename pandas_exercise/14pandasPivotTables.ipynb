```python
import pandas as pd
import seaborn as sns
```

```python
titanic= sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.tail(3)
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

# pivot table ile işlemi yapma

```python
titanic.pivot_table("survived",index="sex", columns="class")
```

```python
age1=pd.cut(titanic["age"],[0,18,90])
age1
```

```python
titanic.pivot_table("survived",["sex",age1],"class")
```

```python
titanic.groupby(["sex",age1,"class"])[["survived"]].aggregate("mean").unstack()
```
