```python
import numpy as np
import pandas as pd
```

```python
a=np.random.randint(7,74,(5,3))
a
```

```python
af=pd.DataFrame(a,columns=["var1","var2","var3"])
af
```

```python
af2=af+99
af2
```

```python
df= pd.concat([af,af2],axis=0)
df
```

# indeks sıralaması düzgün olması için bu arg. ekledik

```python
df2= pd.concat([af,af2],ignore_index=True)
df2
```

```python
df3= pd.concat([af,af2],axis=1)
df3
```

# bu seferde sutun isimlendirme bozukluğu gitmesi için ignore_index arg. kullandık

```python
df3= pd.concat([af,af2],axis=1,ignore_index=True)
df3
```

# farklı sutun isimleri olursa NaN hatası alınır

```python
af2.columns=["var1","var2","value3"]
```

```python
df4=pd.concat([af,af2])
df4
```

## hatayı almamak için iki serinin kesişim bölümlerini aldık

```python
df5=pd.concat([af,af2],join="inner",ignore_index=True)
df5
```
