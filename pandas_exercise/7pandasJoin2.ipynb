```python
import pandas as pd
import numpy as np
```

# birebir birleştirme merge() foksiyonu

```python
df1=pd.DataFrame({"calisanlar":["ali","ahmet","ayse","elif"],"grup":["ik","muhasebe","mühendis","doktor"]})
df1
```

```python
df2=pd.DataFrame({"calisanlar":["ayse","elif","ali","ahmet"],"ilk_giris":[2019,2008,2020,2016]})
df2
```

```python
df3=pd.merge(df1,df2)
df3
```

# many to one join işlemi

```python
df4=pd.DataFrame({"grup":["muhasebe" ,"mühendis","ik","doktor"],"müdür":
                ["caner","yakup","osman","mustafa"]})
df4
```

```python
df5=pd.merge(df3,df4)
df5
```

# many to many join işlemi

```python
df6=pd.DataFrame({"grup":["muhasebe","muhasebe","ik","ik","mühendis","mühendis","doktor","doktor"],
                 "yetenek":["matemetik","exel","kodlama","yönetim","linux","exel","biology","chemistry"]})
df6
```

```python
df7=pd.merge(df5,df6)
df7
```
